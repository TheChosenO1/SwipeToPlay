```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
#All the Columns we Would require
df = pd.read_csv('csvf/Data.csv', usecols=['img', 'title', 'console', 'genre', 'publisher', 'critic_score', 'release_date'])
```

```python
#NaN values in critic square
df.critic_score.isna().sum()
```

```python
L = ['PS3', 'PS4', 'X360', 'XOne', 'PC', 'PS5', 'XS', 'NS']
Curr  = set(L)
```

```python
recent_df = df[df['console'].isin(Curr)]
recent_df = recent_df[recent_df['release_date']>'2007-12-31']
recent_df = recent_df.sort_values('release_date')
```

```python
#First let's combine fames with the same name
df_grouped = recent_df.groupby(['title']).agg({
    'publisher': "first",
    'img': 'first',
    'release_date': 'first',
    'critic_score': lambda x: round(x.mean(), 1),
    'console': lambda x: ', '.join(x)
}).reset_index()
df_grouped
```

```python
for index, row in df_grouped[df_grouped['critic_score'].isna()].iterrows():

    text = ''.join(letter for letter in row['title'] if (letter.isalnum() or letter == ' '))
    driver.get(f"{URL}{text}")
    try:
        c = driver.find_elements(By.XPATH, '//*[@id="__layout"]/div/div[2]/div[1]/div[2]/div[2]/div/a/div[2]/p')[0]
        
        print(c.text)

        c_text = ''.join(letter for letter in c.text if  (letter.isalnum() or letter == ' ')).lower()
        if text.lower() in c_text or c_text in text.lower():
            text =driver.find_elements(By.XPATH , '//*[@id="__layout"]/div/div[2]/div[1]/div[2]/div[2]/div[1]/a/div[3]/div/div/span')[0].text
            if text == 'tbd':
                df_grouped = df_grouped.drop(index)
                print("Drop")
            else:
                df_grouped.loc[index, 'critic_score'] = f"{text[0]}.{text[1]}"
                print(f"{text[0]}.{text[1]}")
        else:
            print(text.lower(), c_text)
            df_grouped = df_grouped.drop(index)
            print("Match Not Found")
    except:
        df_grouped = df_grouped.drop(index)
    
    
df_grouped.to_csv("cleaned.csv")
```
